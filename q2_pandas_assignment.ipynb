```python
import pandas as pd
import numpy as np
from os import listdir
```

```python
failepaths=[f for f in listdir("./") if f.endswith('.csv')]
datafile= pd.concat(map(pd.read_csv,failepaths), ignore_index=True)
failepaths
```

```python
datafile.drop(columns = ['Unnamed: 0'], inplace=True)
```

```python
datafile[['Male','Female']] = datafile.GenderPop.str.split("_",expand=True)
```

```python
datafile.head()
```

```python
datafile.drop(columns = ['GenderPop'], inplace=True)
```

```python
datafile.head()
```

```python
datafile['Income'] = datafile.Income.str.strip('$')
```

```python
datafile.head()
```

```python
datafile=datafile.replace('%','',regex=True)
datafile=datafile.replace('F','',regex=True)
datafile=datafile.replace('M','',regex=True)
```

```python
datafile.head()
```

```python
datafile.loc[:,'Hispanic':'Income'] = round(datafile.loc[:,'Hispanic':'Income'].apply(pd.to_numeric),2)
```

```python
datafile.head()
```

```python
datafile['Male']=datafile['Male'].astype(int)
datafile.dtypes
```

```python
datafile.drop(columns=['Female'],inplace=True)
```

```python
datafile['Female']=datafile['TotalPop']-datafile['Male']
```

```python
datafile.dtypes
```

```python
import matplotlib.pyplot as plt
plt.scatter(datafile['Female'],datafile['Income'])
plt.xlabel('Salary $',fontsize=18)
plt.ylabel('Female_Poplation',fontsize=16)
plt.show()
```

```python
datafile.drop_duplicates(inplace=True)
```

```python
#any() fun print True if there is anyone duplicated record otherwise its print False 
datafile.duplicated().any()
```

```python
plt.scatter(datafile['Female'],datafile['Income'])
plt.xlabel('Salary $',fontsize=18)
plt.ylabel('Female_Poplation',fontsize=16)
plt.show()
```
